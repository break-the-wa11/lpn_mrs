```python
import sys
import os

# Add the parent directory to the system path
sys.path.append(os.path.abspath(os.path.join('..')))
sys.path.append(os.path.abspath(os.path.join('../..')))

from lpn_mrs.datasets import MRSDataset
import torch
import matplotlib.pyplot as plt
```

```python
root = '../data/'
for data_type in ['baseline', 'clean', 'low_lipid']:
    dataset = MRSDataset(root, split='train', data_type=data_type)
    plt.figure(figsize=(10, 5))
    for i in range(len(dataset)):
        sample = dataset[i]
        plt.plot(sample, color = 'red')
    plt.title(f'{data_type} dataset')
    plt.grid()
    plt.show()
```

```python
root = '../data/'
train_dataset = MRSDataset(root, split='train')
train_dataloader = torch.utils.data.DataLoader(
    train_dataset,
    batch_size=100,
    shuffle=True,
    drop_last=True,
    num_workers=8
)
```

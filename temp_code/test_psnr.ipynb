```python
import torch
import numpy as np
import pandas as pd
from lpn_mrs.datasets import MRSDataset
```

$psnr = 10 \log_{10} (range(x)^2 / mse)$.

$mse = range(x)^2 / 10^{psnr / 10}$

```python
dataset = MRSDataset('../data', split='test', data_type='low_lipid')
all_samples = np.array([dataset[i] for i in range(len(dataset))])
for sigma in np.arange(0.01, 0.25, 0.01): 
    noisy = all_samples + np.random.normal(0, sigma, all_samples.shape)

    psnr = np.mean(10 * np.log10((np.max(noisy,axis=1) - np.min(noisy,axis=1))**2 / np.mean((noisy - all_samples)**2, axis=1)))
    print(f'sigma: {sigma:.2f}', psnr)
```

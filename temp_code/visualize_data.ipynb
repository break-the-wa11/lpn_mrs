```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
mrs_path = '../data/train/metab_lipid/Spec_000111.tsv'
data = pd.read_csv(mrs_path, sep='\t')
data.head()
```

```python
print(data.info())
```

```python
# id_min = int(data[(data['ppm'] > 0.2)].index.min())
id_min = 487
id_max = int(data[(data['ppm'] < 4.5)].index.max())
len(data[id_min:id_max])
```

```python
plt.figure(figsize=(12, 6))
plt.plot(data['ppm'], data['FD_Re'], label='Full Data', color='blue', alpha=0.7)
# Plot vertical lines at id_min and id_max
plt.axvline(x=data['ppm'][id_min], color='red', linestyle='--', label=f'id_min (ppm={data["ppm"][id_min]:.2f})')
plt.axvline(x=data['ppm'][id_max], color='green', linestyle='--', label=f'id_max (ppm={data["ppm"][id_max]:.2f})')
plt.xlabel('ppm')
plt.ylabel('FD_Re')
plt.title('FD_Re vs ppm')
plt.grid(True)
print(data['ppm'][id_min],data['ppm'][id_max])
```
